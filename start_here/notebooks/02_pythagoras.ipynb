```python
import math
import numpy as np
import matplotlib.pyplot as plt
import warnings; warnings.filterwarnings("ignore", message=".*distutils Version classes.*", category=DeprecationWarning)
import japanize_matplotlib  # noqa: F401  # 日本語フォント (豆腐化対策)
from ipywidgets import interact, FloatSlider

%matplotlib inline
```

## 1. まず素朴に計算してみる

横3、縦4 の直角三角形の斜辺は？

ピタゴラスの定理:

$$
a^2 + b^2 = c^2 \implies c = \sqrt{a^2 + b^2}
$$

```python
a = 3.0
b = 4.0
c = math.sqrt(a**2 + b**2)
print(f'a = {a}, b = {b}, c = {c}')
```

→ `c = 5.0` になります。3-4-5 の三角形と呼ばれる、4000年前から大工さんが使っている形です。

他の組み合わせも試してみましょう。

```python
for a, b in [(5, 12), (8, 15), (7, 24), (1, 1)]:
    c = math.sqrt(a**2 + b**2)
    print(f'a={a}, b={b}, c={c:.4f}')
```

5-12-13, 8-15-17, 7-24-25 のように、すべて整数になる組み合わせを **ピタゴラス数** と呼びます。

最後の 1-1-?√2 だけは整数になりません。これが古代ギリシャ人を震撼させた **無理数** の登場です。

## 2. 絵で見る

直角三角形と、その辺の上に正方形を描いてみます。「斜辺の上の正方形の面積 = 他2辺の正方形の面積の和」になります。

青の正方形 ($a^2$) + 緑の正方形 ($b^2$) = 黄の正方形 ($c^2$)

```python
def draw_pythagoras(a: float, b: float) -> None:
    """直角三角形と3つの正方形を描く."""
    c = math.sqrt(a**2 + b**2)

    fig, ax = plt.subplots(figsize=(8, 8))

    # 三角形
    triangle = plt.Polygon([[0, 0], [a, 0], [0, b]], fill=False, edgecolor='black', linewidth=2)
    ax.add_patch(triangle)

    # 底辺 a の上の正方形
    sq_a = plt.Rectangle((0, -a), a, a, fill=True, facecolor='lightblue', alpha=0.5, edgecolor='blue')
    ax.add_patch(sq_a)
    ax.text(a / 2, -a / 2, f'a² = {a**2:.1f}', ha='center', va='center', fontsize=12)

    # 縦辺 b の左の正方形
    sq_b = plt.Rectangle((-b, 0), b, b, fill=True, facecolor='lightgreen', alpha=0.5, edgecolor='green')
    ax.add_patch(sq_b)
    ax.text(-b / 2, b / 2, f'b² = {b**2:.1f}', ha='center', va='center', fontsize=12)

    # 斜辺の上の正方形（回転）
    angle = math.atan2(b, -a)  # 斜辺の方向
    cos_a, sin_a = math.cos(angle), math.sin(angle)
    # 斜辺の両端: (a, 0) と (0, b)
    # 斜辺の外側に正方形を作る
    dx = -b / c
    dy = -a / c
    p1 = np.array([a, 0])
    p2 = np.array([0, b])
    p3 = p2 + c * np.array([dx, dy])
    p4 = p1 + c * np.array([dx, dy])
    sq_c = plt.Polygon([p1, p2, p3, p4], fill=True, facecolor='lightyellow', alpha=0.5, edgecolor='orange')
    ax.add_patch(sq_c)
    cx, cy = (p1 + p3) / 2
    ax.text(cx, cy, f'c² = {c**2:.1f}', ha='center', va='center', fontsize=12)

    ax.set_xlim(-b - 1, a + 1)
    ax.set_ylim(-a - 1, b + 1)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    ax.set_title(f'a² + b² = c²  →  {a**2:.1f} + {b**2:.1f} = {c**2:.1f}')
    plt.show()

draw_pythagoras(3, 4)
```

黄色の正方形 (c²) と、青と緑の正方形 (a² + b²) の **面積が等しい**。これがピタゴラスの定理の幾何学的な意味です。

## 3. スライダーで遊ぶ

下のセルを実行すると、a と b をスライダーで動かせます。

```python
interact(
    draw_pythagoras,
    a=FloatSlider(min=0.5, max=8.0, step=0.5, value=3.0, description='a'),
    b=FloatSlider(min=0.5, max=8.0, step=0.5, value=4.0, description='b'),
);
```

## 4. 実生活への応用: 地図上の距離

「家から駅まで、まっすぐ最短ルートで何メートル？」を計算してみます。

東京駅(原点)から、東に 800m、北に 600m の場所にあなたがいるとします。直線距離は？

地図上の距離 (ユークリッド距離):

$$
d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
$$

```python
east = 800  # 東に何メートル (m)
north = 600  # 北に何メートル (m)

distance = math.sqrt(east**2 + north**2)
print(f'直線距離: {distance:.1f} メートル')
```

→ 1000 メートル (= 1 km)。これは GPS が距離を計算するときに使う基本式と同じです。

## 5. AI への接続: 「データの類似度」

ピタゴラスの定理を高次元に拡張すると、AIで使う **ユークリッド距離** になります。

2人のユーザーの「映画の評価ベクトル」を比べてみましょう (5段階評価)。

ユークリッド距離 (高次元):

$$
d(\mathbf{u}, \mathbf{v}) = \sqrt{\sum_i (u_i - v_i)^2}
$$

```python
# 5つの映画への評価
user_a = np.array([5, 3, 4, 4, 2])  # ユーザーA
user_b = np.array([4, 3, 5, 5, 1])  # ユーザーB
user_c = np.array([1, 5, 1, 2, 5])  # ユーザーC

def euclidean_distance(u: np.ndarray, v: np.ndarray) -> float:
    """高次元ピタゴラスの定理 = ユークリッド距離."""
    return float(np.sqrt(np.sum((u - v) ** 2)))

print(f'A vs B: {euclidean_distance(user_a, user_b):.4f}')
print(f'A vs C: {euclidean_distance(user_a, user_c):.4f}')
print(f'→ A と B のほうが「近い」ので、A に B のおすすめを提案する、というのが推薦システムの基本')
```

## まとめ

- `a² + b² = c²` という素朴な式が、**地図・GPS・AI** すべてを支えている
- 「2点間の直線距離」を計算する道具として、無数の場面で使われている
- これがピタゴラスの定理の現代的な意義
