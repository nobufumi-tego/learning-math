```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch
from ipywidgets import interact, FloatSlider

%matplotlib inline
```

## 1. sin / cos は「単位円の上の点の位置」

半径1の円 (= 単位円) を考えて、角度 θ だけ反時計回りに回ったときの点の座標が `(cos θ, sin θ)` です。

```python
def draw_unit_circle(theta_deg: float) -> None:
    """単位円上の点と sin/cos を描く.

    Args:
        theta_deg: 角度 (度)
    """
    theta = math.radians(theta_deg)
    x = math.cos(theta)
    y = math.sin(theta)

    fig, ax = plt.subplots(figsize=(7, 7))
    circle = plt.Circle((0, 0), 1, fill=False, color='gray', linewidth=2)
    ax.add_patch(circle)

    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)

    # 点
    ax.plot(x, y, 'ro', markersize=12)
    ax.plot([0, x], [0, y], 'r-', linewidth=2)

    # cos (横方向)
    ax.plot([0, x], [0, 0], 'b-', linewidth=3, label=f'cos θ = {x:.3f}')
    ax.plot([x, x], [0, y], 'g-', linewidth=3, label=f'sin θ = {y:.3f}')

    ax.set_xlim(-1.3, 1.3)
    ax.set_ylim(-1.3, 1.3)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    ax.legend(loc='upper left', fontsize=11)
    ax.set_title(f'θ = {theta_deg}° の単位円上の点')
    plt.show()

draw_unit_circle(60)
```

**スライダーで動かして体感**:

```python
interact(
    draw_unit_circle,
    theta_deg=FloatSlider(min=0, max=360, step=5, value=60, description='角度 (度)'),
);
```

## 2. 円運動 → 波

θ を増やしていったときの sin の値を時間軸でプロットすると、**波** になります。これが「音」「電波」「光」の基本です。

```python
thetas = np.linspace(0, 4 * np.pi, 500)  # 0 〜 4π ラジアン
sin_values = np.sin(thetas)
cos_values = np.cos(thetas)

fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(thetas, sin_values, 'g-', label='sin θ', linewidth=2)
ax.plot(thetas, cos_values, 'b-', label='cos θ', linewidth=2)
ax.axhline(0, color='black', linewidth=0.5)
ax.set_xlabel('θ (ラジアン)')
ax.set_ylabel('値')
ax.set_title('sin と cos は、ぴったり 90° (π/2) ずれた波')
ax.legend(fontsize=12)
ax.grid(True, alpha=0.3)
plt.show()
```

## 3. 波を合成すると音色になる

音は sin の波。複数の sin を足し合わせると「ハーモニー」「音色」が作れます。

```python
def plot_wave_sum(f1: float, f2: float) -> None:
    """周波数 f1 と f2 の波を合成して表示."""
    t = np.linspace(0, 1, 1000)   # 1秒分
    w1 = np.sin(2 * np.pi * f1 * t)
    w2 = np.sin(2 * np.pi * f2 * t)
    w_sum = w1 + w2

    fig, axes = plt.subplots(3, 1, figsize=(12, 6), sharex=True)
    axes[0].plot(t, w1, 'b-')
    axes[0].set_title(f'波1: {f1} Hz')
    axes[1].plot(t, w2, 'g-')
    axes[1].set_title(f'波2: {f2} Hz')
    axes[2].plot(t, w_sum, 'r-')
    axes[2].set_title(f'合成: 波1 + 波2')
    axes[2].set_xlabel('時間 (秒)')
    for ax in axes:
        ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

plot_wave_sum(5, 7)
```

スライダーで自分で波を合成してみよう:

```python
interact(
    plot_wave_sum,
    f1=FloatSlider(min=1, max=20, step=1, value=3, description='周波数1 (Hz)'),
    f2=FloatSlider(min=1, max=20, step=1, value=5, description='周波数2 (Hz)'),
);
```

周波数が近いと「うなり (beat)」が発生し、ピアノの調律師はこれを聞き取って音を合わせます。

## 4. ビルの高さを測る (三角測量)

30m 離れた場所からビルのてっぺんを 60° の角度で見上げる。ビルの高さは？

`tan 60° = 高さ / 30` なので、高さ = `30 × tan 60°`

```python
def building_height(distance: float, angle_deg: float) -> float:
    """見上げ角度と距離からビルの高さを計算.

    Args:
        distance: ビルからの水平距離 (m)
        angle_deg: 仰角 (度)

    Returns:
        ビルの高さ (m)
    """
    angle = math.radians(angle_deg)
    return distance * math.tan(angle)

for angle in [30, 45, 60, 75]:
    h = building_height(30, angle)
    print(f'30m離れて {angle}度 で見上げる → 高さ {h:.2f} m')
```

```python
def draw_building(distance: float, angle_deg: float) -> None:
    """三角測量を絵で."""
    h = building_height(distance, angle_deg)
    fig, ax = plt.subplots(figsize=(10, 6))
    # 地面
    ax.plot([-5, distance + 5], [0, 0], 'k-', linewidth=2)
    # 人
    ax.plot(0, 0.5, 'bo', markersize=12)
    ax.text(0, -1.5, '👤 観測者', ha='center', fontsize=11)
    # ビル
    ax.add_patch(plt.Rectangle((distance - 1, 0), 2, h, color='lightgray', edgecolor='black'))
    ax.text(distance, h / 2, '🏢', ha='center', va='center', fontsize=20)
    # 視線
    ax.plot([0, distance], [0.5, h], 'r--', linewidth=1.5)
    # 角度表記
    ax.text(2, 1, f'{angle_deg}°', color='red', fontsize=12)
    # 距離表記
    ax.annotate('', xy=(distance, -2), xytext=(0, -2), arrowprops=dict(arrowstyle='<->'))
    ax.text(distance / 2, -3, f'{distance} m', ha='center', fontsize=11)
    # 高さ表記
    ax.annotate('', xy=(distance + 4, h), xytext=(distance + 4, 0), arrowprops=dict(arrowstyle='<->'))
    ax.text(distance + 5, h / 2, f'{h:.1f} m', fontsize=11)

    ax.set_xlim(-5, distance + 10)
    ax.set_ylim(-5, max(h + 5, 20))
    ax.set_aspect('equal')
    ax.set_title(f'三角測量: 距離 {distance}m + 仰角 {angle_deg}° → 高さ {h:.2f}m')
    ax.grid(True, alpha=0.3)
    plt.show()

interact(
    draw_building,
    distance=FloatSlider(min=10, max=100, step=5, value=30, description='距離 (m)'),
    angle_deg=FloatSlider(min=10, max=85, step=1, value=60, description='仰角 (度)'),
);
```

## まとめ

- sin / cos は **単位円の上の点の (x, y) 座標**
- これを時間軸で見ると **波** になる → 音・光・電波の本質
- tan は **傾き** → 三角測量で「測れない高さ」を測れる
