```python
import itertools

def implies(p, q):
    """p ⇒ q を計算."""
    return (not p) or q

def equiv(p, q):
    """p ⇔ q."""
    return p == q
```

## 1. 基本演算の真理表

$\neg P$, $P \land Q$, $P \lor Q$, $P \Rightarrow Q$, $P \Leftrightarrow Q$

```python
print(f'{"P":^6}{"Q":^6}{"¬P":^6}{"P∧Q":^6}{"P∨Q":^6}{"P⇒Q":^6}{"P⇔Q":^6}')
print('-' * 42)
for P in [True, False]:
    for Q in [True, False]:
        print(f'{str(P)[0]:^6}{str(Q)[0]:^6}{str(not P)[0]:^6}{str(P and Q)[0]:^6}{str(P or Q)[0]:^6}{str(implies(P,Q))[0]:^6}{str(equiv(P,Q))[0]:^6}')
```

## 2. ド・モルガンの法則

$$
\neg(P \land Q) \Leftrightarrow \neg P \lor \neg Q
$$

$$
\neg(P \lor Q) \Leftrightarrow \neg P \land \neg Q
$$

```python
def truth_table(formula1, formula2, var_names):
    """2 つの式が論理的に同値か真理表で検証."""
    print(' | '.join(var_names) + ' || formula1 || formula2 || 一致?')
    print('-' * 50)
    all_equal = True
    for vals in itertools.product([True, False], repeat=len(var_names)):
        env = dict(zip(var_names, vals))
        r1 = formula1(**env)
        r2 = formula2(**env)
        same = r1 == r2
        all_equal = all_equal and same
        vals_str = ' | '.join(f'{str(v)[0]:>5}' for v in vals)
        print(f'{vals_str} || {str(r1)[0]:>8} || {str(r2)[0]:>8} || {"✅" if same else "❌"}')
    print(f'\n=> 論理的同値: {all_equal}')

print('=== ド・モルガン (1): ¬(P∧Q) ⇔ ¬P∨¬Q ===')
truth_table(
    lambda P, Q: not (P and Q),
    lambda P, Q: (not P) or (not Q),
    ['P', 'Q'],
)

print('\n=== ド・モルガン (2): ¬(P∨Q) ⇔ ¬P∧¬Q ===')
truth_table(
    lambda P, Q: not (P or Q),
    lambda P, Q: (not P) and (not Q),
    ['P', 'Q'],
)
```

## 3. 含意の対偶

$$
(P \Rightarrow Q) \Leftrightarrow (\neg Q \Rightarrow \neg P)
$$

「P ならば Q」と「Q でないなら P でない」は論理的に同値

```python
print('=== 含意 ⇔ 対偶 ===')
truth_table(
    lambda P, Q: implies(P, Q),
    lambda P, Q: implies(not Q, not P),
    ['P', 'Q'],
)
```

## 4. 含意 ⇔ ¬P ∨ Q

$$
(P \Rightarrow Q) \Leftrightarrow (\neg P \lor Q)
$$

```python
truth_table(
    lambda P, Q: implies(P, Q),
    lambda P, Q: (not P) or Q,
    ['P', 'Q'],
)
```

## 5. 排中律と矛盾律

$P \lor \neg P$ (排中律): すべてのケースで真 → 恒真式

$P \land \neg P$ (矛盾律): すべてのケースで偽 → 矛盾

```python
print('排中律 P ∨ ¬P:')
for P in [True, False]:
    print(f'  P={str(P)[0]}: {P or not P}')

print('\n矛盾律 P ∧ ¬P:')
for P in [True, False]:
    print(f'  P={str(P)[0]}: {P and not P}')
```

## まとめ
- 真理表で論理式の評価が完全に機械的にできる
- ド・モルガンの法則・対偶などの恒等式を Python で検証可
- 「含意」が日常会話と違う点 (偽 ⇒ 何でも = 真) を実体感
