## 1. 直接証明: 偶数の 2 乗は偶数

$$
n \text{ 偶数} \implies n^2 \text{ も偶数}
$$

証明: $n = 2k$ なら $n^2 = 4k^2 = 2 \cdot (2k^2)$ → 偶数

```python
print('偶数の 2 乗が偶数か確認 (シミュレーション):')
for n in range(2, 21, 2):
    print(f'  {n:>2}² = {n**2:>4}  偶数: {n**2 % 2 == 0}')

# 大きい数でもチェック
print('\n大きい偶数でも:')
for n in [1000, 12_345 * 2, 1_000_000_002]:
    print(f'  {n:>15}² = {n**2:>30}  偶数: {n**2 % 2 == 0}')
```

## 2. 背理法: $\sqrt{2}$ は無理数

$$
\sqrt{2} \notin \mathbb{Q}
$$

証明: $\sqrt{2} = a/b$ (既約分数) と仮定 → $a, b$ がともに偶数になり矛盾

シミュレーション: 多くの分数を試して $\sqrt{2}$ に一致するか調べる (**証明ではない**)

```python
sqrt2 = np.sqrt(2)
print(f'√2 ≈ {sqrt2}')
print()

# 様々な分数で近似
approximations = [
    (1, 1), (3, 2), (7, 5), (17, 12), (41, 29), (99, 70), (577, 408)
]
for num, denom in approximations:
    val = num / denom
    err = abs(val - sqrt2)
    print(f'  {num:>4}/{denom:<4} = {val:.10f}  (誤差 {err:.2e})')

print('\n→ どんなに精度を上げても "完全一致" にはならない (無理数だから)')
```

## 3. 数学的帰納法: $\sum_{i=1}^n i = \frac{n(n+1)}{2}$

ガウスが小学生のとき発見したと言われる公式

```python
def sum_formula(n):
    """n(n+1)/2."""
    return n * (n + 1) // 2

print(f'{"n":>10} | {"直接和":>20} | {"公式":>20} | 一致')
print('-' * 65)
for n in [1, 5, 10, 100, 1_000, 1_000_000]:
    direct = sum(range(1, n + 1))
    formula = sum_formula(n)
    print(f'{n:>10} | {direct:>20} | {formula:>20} | {"✅" if direct == formula else "❌"}')
```

```python
# 視覚的に: 三角数として表現
fig, axes = plt.subplots(1, 5, figsize=(15, 4))
for ax, n in zip(axes, [1, 2, 3, 4, 5]):
    pts_x, pts_y = [], []
    for i in range(1, n + 1):
        for j in range(i):
            pts_x.append(j)
            pts_y.append(-i)
    ax.scatter(pts_x, pts_y, s=200, c='blue')
    ax.set_title(f'n={n}, 和={sum_formula(n)}')
    ax.set_xlim(-1, n + 1); ax.set_ylim(-n - 1, 0)
    ax.set_aspect('equal'); ax.axis('off')
plt.tight_layout()
plt.show()

print('1+2+3+...+n = n(n+1)/2  ← 三角形の点の数')
```

## 4. 別の帰納法: $1^2 + 2^2 + \ldots + n^2 = \frac{n(n+1)(2n+1)}{6}$

平方和の公式

```python
def square_sum_formula(n):
    return n * (n + 1) * (2 * n + 1) // 6

print(f'{"n":>10} | {"直接和":>15} | {"公式":>15} | 一致')
for n in [1, 5, 10, 100, 1000]:
    direct = sum(i**2 for i in range(1, n + 1))
    formula = square_sum_formula(n)
    print(f'{n:>10} | {direct:>15} | {formula:>15} | {"✅" if direct == formula else "❌"}')
```

## 5. 帰納法と再帰関数 — 「証明可能性」の対応

$$
\text{factorial}(n) = \begin{cases} 1 & n = 0 \\ n \cdot \text{factorial}(n-1) & n > 0 \end{cases}
$$

$\text{factorial}(n) = n!$ を帰納法で証明できる:
- 基底: $\text{factorial}(0) = 1 = 0!$ ✅
- 帰納: $\text{factorial}(k) = k!$ なら $\text{factorial}(k+1) = (k+1) \cdot k! = (k+1)!$

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

import math
for n in range(10):
    print(f'factorial({n}) = {factorial(n):>10},  math.factorial({n}) = {math.factorial(n):>10}')
```
