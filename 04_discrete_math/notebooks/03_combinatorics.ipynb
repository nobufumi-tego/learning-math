```python
import math
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import warnings; warnings.filterwarnings("ignore", message=".*distutils Version classes.*", category=DeprecationWarning)
import japanize_matplotlib  # noqa: F401  # 日本語フォント (豆腐化対策)
from itertools import permutations, combinations, combinations_with_replacement

%matplotlib inline
```

## 1. 階乗 $n!$

$$
n! = n \cdot (n-1) \cdots 2 \cdot 1, \quad 0! = 1
$$

```python
for n in [0, 1, 2, 5, 10, 20]:
    print(f'{n}! = {math.factorial(n):,}')
```

## 2. 順列 $P(n, r)$

$$
P(n, r) = \frac{n!}{(n-r)!}
$$

「n 個から r 個選んで**並べる**」

```python
n, r = 5, 3
print(f'P({n}, {r}) = {math.perm(n, r)}')

# 全列挙
perms = list(permutations(['A', 'B', 'C', 'D', 'E'], r))
print(f'実際に列挙した数: {len(perms)}')
print(f'最初の 6 個: {perms[:6]}')
```

## 3. 組合せ $\binom{n}{r}$

$$
\binom{n}{r} = \frac{n!}{r!(n-r)!} = \frac{P(n, r)}{r!}
$$

「n 個から r 個**選ぶ**」(順序問わず)

```python
n, r = 5, 3
print(f'C({n}, {r}) = {math.comb(n, r)}')

combs = list(combinations(['A', 'B', 'C', 'D', 'E'], r))
print(f'実際に列挙した数: {len(combs)}')
print(f'すべて: {combs}')
```

## 4. パスカルの三角形

$$
\binom{n}{r} = \binom{n-1}{r-1} + \binom{n-1}{r}
$$

```python
def pascal(n):
    return [math.comb(n, r) for r in range(n + 1)]

print('パスカルの三角形 (n=0〜10):')
max_width = len(' '.join(f'{v:>4}' for v in pascal(10)))
for n in range(11):
    row = pascal(n)
    line = ' '.join(f'{v:>4}' for v in row)
    print(line.center(max_width))
```

## 5. 二項定理

$$
(x + y)^n = \sum_{r=0}^n \binom{n}{r} x^{n-r} y^r
$$

```python
x, y = sp.symbols('x y')
for n in range(1, 6):
    expanded = sp.expand((x + y)**n)
    print(f'(x + y)^{n} = {expanded}')
```

## 6. 二項分布との接続

$$
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
$$

コインを 10 回投げて表が k 回出る確率

```python
n, p = 10, 0.5
ks = list(range(n + 1))
probs = [math.comb(n, k) * p**k * (1-p)**(n-k) for k in ks]

fig, ax = plt.subplots(figsize=(10, 4))
ax.bar(ks, probs, alpha=0.7, edgecolor='black')
ax.set_xlabel('k (表の回数)')
ax.set_ylabel('P(X = k)')
ax.set_title(f'$\\mathrm{{Bin}}(n={n}, p={p})$:  $P(X=k) = \\binom{{n}}{{k}} p^k (1-p)^{{n-k}}$')
ax.grid(True, alpha=0.3)
plt.show()

for k, prob in zip(ks, probs):
    bar = '█' * int(prob * 100)
    print(f'P(X={k:>2}) = {prob:.4f}  {bar}')

print(f'\n合計: {sum(probs):.6f}  (理論 1.0)')
```

## 7. スターリングの公式

$$
n! \approx \sqrt{2\pi n} \left(\frac{n}{e}\right)^n
$$

```python
def stirling(n):
    return math.sqrt(2 * math.pi * n) * (n / math.e) ** n

print(f'{"n":>5} | {"n! (実際)":>20} | {"スターリング近似":>20} | 比率')
print('-' * 65)
for n in [5, 10, 20, 50, 100]:
    actual = math.factorial(n)
    approx = stirling(n)
    print(f'{n:>5} | {actual:>20.4e} | {approx:>20.4e} | {approx/actual:.6f}')
print('\n→ n が大きくなるほど比率が 1 に近づく (近似精度↑)')
```
