## 1. 等号 `=` と 等価 `==` の違い

数学の `=` は「等しい」 を主張する。Python では:
- **代入** には `=` (左辺に値を入れる)
- **等価判定** には `==` (Bool を返す)

```python
x = 5            # 代入: x に 5 を入れる
print(x == 5)    # 等価判定: True
print(x == 6)    # False
print(x != 6)    # ≠ (not equal): True
```

## 2. 不等号 `<`, `≤`, `>`, `≥`

| 数学記号 | Python | 意味 |
|---|---|---|
| $a < b$ | `a < b` | 小なり |
| $a \leq b$ | `a <= b` | 以下 |
| $a > b$ | `a > b` | 大なり |
| $a \geq b$ | `a >= b` | 以上 |

```python
a, b = 3, 5
print(f'a < b:  {a < b}')
print(f'a <= b: {a <= b}')
print(f'a > b:  {a > b}')
print(f'a >= b: {a >= b}')

# 連続不等式 (Python 特有の機能)
print(f'1 < a < 10: {1 < a < 10}')  # 数学そのままに書ける
```

## 3. 無限大 `∞`

数学の $\infty$ は概念だが、Python では `math.inf` (または `float('inf')`) で扱える。

```python
inf = math.inf
print(f'inf > 10**100:  {inf > 10**100}')
print(f'inf + 1 == inf: {inf + 1 == inf}')
print(f'1 / inf:        {1 / inf}')
print(f'-inf:            {-inf}')
print(f'inf - inf:       {inf - inf}')  # NaN: 数学的に不定
print(f'np.inf == math.inf: {np.inf == math.inf}')
```

## 4. `±` (プラスマイナス)

数学では「 $a = 3 \pm 0.5$ 」 のように両方の値を意味する。Python ではタプルや配列で表現する。

```python
a = 3
delta = 0.5
values = (a - delta, a + delta)
print(f'a ± δ = {values}  → {values[0]} と {values[1]}')

# 二次方程式の解の公式: x = (-b ± √(b²-4ac)) / 2a
import math
b, c, a_coef = -3, 2, 1
discriminant = b**2 - 4*a_coef*c
x_plus  = (-b + math.sqrt(discriminant)) / (2*a_coef)
x_minus = (-b - math.sqrt(discriminant)) / (2*a_coef)
print(f'x² - 3x + 2 = 0 の解: x = {x_plus} または {x_minus}')
```

## 5. ≈ (ニアリーイコール)

浮動小数点演算は誤差を含むので、`==` で厳密比較すると失敗することが多い。`math.isclose()` か `np.isclose()` を使う。

```python
x = 0.1 + 0.2
print(f'0.1 + 0.2 = {x}')
print(f'== 0.3: {x == 0.3}')      # False! (浮動小数点誤差)
print(f'isclose: {math.isclose(x, 0.3)}')  # True (誤差を許容)
```
