```python
import math
import numpy as np
import matplotlib.pyplot as plt
import warnings; warnings.filterwarnings("ignore", message=".*distutils Version classes.*", category=DeprecationWarning)
import japanize_matplotlib  # noqa: F401  # 日本語フォント (豆腐化対策)

%matplotlib inline
```

## 1. 数学の $f(x)$ を Python で書く

例: $f(x) = x^2 + 1$

```python
# (a) 関数定義
def f(x: float) -> float:
    return x**2 + 1

# (b) lambda 式 (匿名関数)
f_lambda = lambda x: x**2 + 1

print(f'f(3) = {f(3)}')
print(f'f_lambda(3) = {f_lambda(3)}')
print(f'f(-2) = {f(-2)}')
```

## 2. 写像の矢印 `↦`

数学では $f: \mathbb{R} \to \mathbb{R}, \; x \mapsto x^2 + 1$ と書く。

- `→` (`A → B`): どの集合からどの集合への写像か (型シグネチャ)
- `↦` (`x ↦ f(x)`): 入力 x を出力 f(x) に「移す」

Python の **型ヒント** はこれに相当する: `def f(x: float) -> float` の `-> float` 部分。

## 3. グラフを描く

$f(x) = x^2 + 1$ のグラフを matplotlib で。

```python
xs = np.linspace(-3, 3, 100)
ys = f(xs)  # NumPy 配列に対しても一発で動く

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(xs, ys, linewidth=2)
ax.set_title('$f(x) = x^2 + 1$')
ax.set_xlabel('x (定義域 ℝ)')
ax.set_ylabel('f(x) (値域 [1, ∞))')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 4. 合成関数 $f \circ g$

$(f \circ g)(x) = f(g(x))$ という入れ子。

```python
def g(x: float) -> float:
    return 2*x + 3

def compose(f, g):
    '''f ∘ g を返す.'''
    return lambda x: f(g(x))

f_g = compose(f, g)        # f∘g
g_f = compose(g, f)        # g∘f

print(f'g(2) = {g(2)}  → f(g(2)) = {f(g(2))}')
print(f'(f ∘ g)(2) = {f_g(2)}')
print(f'(g ∘ f)(2) = {g_f(2)}')
print(f'\n合成は順序を変えると結果が変わる: f∘g ≠ g∘f')
```

## 5. 逆関数 $f^{-1}$

$y = f(x)$ を $x$ について解いて入れ替える。
例: $f(x) = 2x + 3$ の逆関数は $f^{-1}(y) = (y - 3) / 2$。

```python
def h(x): return 2*x + 3
def h_inv(y): return (y - 3) / 2

# 検証: h(h_inv(y)) == y
for y in [0, 5, -1, 100]:
    x = h_inv(y)
    y_back = h(x)
    assert math.isclose(y, y_back)
print('✅ h(h_inv(y)) = y が全てのテスト値で成立')

print(f'\nh(5) = {h(5)}')
print(f'h_inv(13) = {h_inv(13)}  ← h(5) の逆は 5')
```

## 6. 多変数関数 $f: \mathbb{R}^2 \to \mathbb{R}$

機械学習では数千次元の入力が普通。まずは 2 変数で雰囲気を。

```python
def quad(x: float, y: float) -> float:
    '''2 変数二次関数.'''
    return x**2 + y**2

# 等高線で可視化
xs = np.linspace(-3, 3, 50)
ys = np.linspace(-3, 3, 50)
X, Y = np.meshgrid(xs, ys)
Z = quad(X, Y)

fig, ax = plt.subplots(figsize=(6, 5))
cf = ax.contourf(X, Y, Z, levels=15, cmap='viridis')
ax.contour(X, Y, Z, levels=15, colors='white', linewidths=0.5)
ax.set_title('$f(x, y) = x^2 + y^2$')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.colorbar(cf, ax=ax, label='f(x, y)')
plt.tight_layout()
plt.show()
```

## ⚠️ よくあるエラー — 文字列を直接 int に変換

`int()` は **整数の文字列** しか直接変換できません。小数の文字列なら float を経由する必要があります。

```python
# 試してみよう: 小数の文字列を int に
try:
    n = int("3.5")
except ValueError as e:
    print(f'❌ {type(e).__name__}: {e}')
    print('  → int() は整数の文字列しか直接受け取れない')
    print()
    print('  → 小数を経由すれば OK (切り捨て):')
    print(f'  int(float("3.5")) = {int(float("3.5"))}')
    print(f'  round(float("3.5")) = {round(float("3.5"))}  ← 偶数丸め')

# エラー読解の詳細: start_here/00_pet_terminal/columns/05_reading_errors.md
```
