## 1. ∧ (and), ∨ (or), ¬ (not)

| 数学 | Python |
|---|---|
| $P \land Q$ | `P and Q` |
| $P \lor Q$ | `P or Q` |
| $\lnot P$ | `not P` |

```python
P, Q = True, False
print(f'P ∧ Q: {P and Q}')
print(f'P ∨ Q: {P or Q}')
print(f'¬P:    {not P}')
print(f'¬Q:    {not Q}')
```

## 2. ∀ (全称) と ∃ (存在)

- $\forall x \in A, P(x)$ → 「**全ての** x で P が真」 → `all(P(x) for x in A)`
- $\exists x \in A, P(x)$ → 「**ある** x で P が真」 → `any(P(x) for x in A)`

```python
numbers = [2, 4, 6, 8, 10]

# ∀ x ∈ numbers, x は偶数か?
all_even = all(x % 2 == 0 for x in numbers)
print(f'∀ x ∈ {numbers}, x は偶数: {all_even}')

# ∃ x ∈ numbers, x > 5 か?
exists_gt5 = any(x > 5 for x in numbers)
print(f'∃ x ∈ {numbers}, x > 5: {exists_gt5}')

# ∀ と ∃ は対 (∀P = ¬∃¬P)
nums2 = [1, 3, 5, 7, 8]
print(f'\n{nums2} は全て奇数か? {all(x % 2 == 1 for x in nums2)}')  # False (8 がいる)
```

## 3. 含意 `⇒` (implies)

$P \Rightarrow Q$ は **「P が真なら Q も真」** という主張。

真理値表:

| P | Q | P ⇒ Q |
|---|---|---|
| T | T | T |
| T | F | **F** |
| F | T | T |
| F | F | T |

Python 標準には `⇒` 演算子は無いので、論理同値 $P \Rightarrow Q \equiv \lnot P \lor Q$ を使って実装。

```python
def implies(p: bool, q: bool) -> bool:
    '''P ⇒ Q の論理演算 (≡ ¬P ∨ Q).'''
    return (not p) or q

# 真理値表を表示
print('P | Q | P ⇒ Q')
print('--+---+------')
for p in [True, False]:
    for q in [True, False]:
        print(f'{str(p)[0]} | {str(q)[0]} | {str(implies(p, q))[0]}')
```

## 4. 同値 `⇔` (iff = if and only if)

$P \Leftrightarrow Q$ は「P と Q が同じ真偽値」。

```python
def iff(p: bool, q: bool) -> bool:
    '''P ⇔ Q.'''
    return p == q

print('P | Q | P ⇔ Q')
print('--+---+------')
for p in [True, False]:
    for q in [True, False]:
        print(f'{str(p)[0]} | {str(q)[0]} | {str(iff(p, q))[0]}')
```

## 5. ド・モルガンの法則を検証

$\lnot (P \land Q) \equiv (\lnot P) \lor (\lnot Q)$
$\lnot (P \lor Q) \equiv (\lnot P) \land (\lnot Q)$

```python
for p in [True, False]:
    for q in [True, False]:
        lhs1 = not (p and q)
        rhs1 = (not p) or (not q)
        lhs2 = not (p or q)
        rhs2 = (not p) and (not q)
        assert lhs1 == rhs1
        assert lhs2 == rhs2
print('✅ ド・モルガンの法則は全パターンで成立')
```
