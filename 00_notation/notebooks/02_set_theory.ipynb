## 1. 集合の作り方

数学の $A = \{1, 2, 3\}$ は Python では `set` で。

```python
A = {1, 2, 3}
B = {2, 3, 4, 5}
empty = set()    # 空集合 ∅ ({} は dict なので注意)

print(f'A = {A}')
print(f'B = {B}')
print(f'∅ = {empty}')
print(f'|A| = {len(A)}  (集合の濃度)')
```

## 2. 帰属 `∈` / 非帰属 `∉`

| 数学 | Python |
|---|---|
| $x \in A$ | `x in A` |
| $x \notin A$ | `x not in A` |

```python
print(f'2 ∈ A:  {2 in A}')
print(f'5 ∈ A:  {5 in A}')
print(f'5 ∉ A:  {5 not in A}')
```

## 3. 部分集合 `⊆`, `⊂`

- `⊆` (部分集合 — 等しくてもよい): `A.issubset(B)` または `A <= B`
- `⊂` (真部分集合 — 等しくない): `A < B`

```python
C = {1, 2}
print(f'C ⊆ A: {C.issubset(A)}')
print(f'C ⊂ A: {C < A}')         # 真部分集合
print(f'A ⊆ A: {A <= A}')         # 部分集合 (等号含む)
print(f'A ⊂ A: {A < A}')          # 真部分集合 ではない
```

## 4. 和集合 `∪` / 積集合 `∩` / 差集合 `∖`

| 数学 | Python 演算子 | メソッド |
|---|---|---|
| $A \cup B$ | `A \| B` | `A.union(B)` |
| $A \cap B$ | `A & B` | `A.intersection(B)` |
| $A \setminus B$ | `A - B` | `A.difference(B)` |
| $A \triangle B$ (対称差) | `A ^ B` | `A.symmetric_difference(B)` |

```python
print(f'A ∪ B = {A | B}')
print(f'A ∩ B = {A & B}')
print(f'A ∖ B = {A - B}')
print(f'A △ B = {A ^ B}')
```

## 5. 集合の応用 — 重複削除

機械学習や統計でデータの「ユニーク値」 を取るのに集合は便利。

```python
data = [1, 2, 2, 3, 4, 4, 4, 5]
unique = set(data)
print(f'元データ: {data}')
print(f'ユニーク: {unique}  ({len(unique)} 種類)')

# 2 つのデータの共通要素 / 片方だけにある要素
users_a = {'alice', 'bob', 'carol', 'dave'}
users_b = {'carol', 'dave', 'eve', 'frank'}
print(f'両方を使った人: {users_a & users_b}')
print(f'A だけ:        {users_a - users_b}')
print(f'どちらかは使った: {users_a | users_b}')
```

## 6. 直積集合 $A \times B$

数学では順序対 $(a, b)$ の集合。Python では `itertools.product`。

```python
from itertools import product
X = {1, 2}
Y = {'a', 'b'}
pairs = set(product(X, Y))
print(f'X × Y = {pairs}')
```
