## 1. 主要ギリシャ文字一覧と慣習的用法

| 文字 | 名前 | 主な用法 |
|---|---|---|
| $\alpha$ | アルファ | 学習率、係数、有意水準 |
| $\beta$ | ベータ | 係数、運動量 (Adam) |
| $\gamma$ | ガンマ | 学習率減衰、相対関係 |
| $\delta$ | デルタ | 微小量、デルタ関数 |
| $\epsilon, \varepsilon$ | イプシロン | 微小量、許容誤差、探索率 |
| $\theta$ | シータ | パラメータ全般、角度 |
| $\lambda$ | ラムダ | 正則化係数、固有値、波長 |
| $\mu$ | ミュー | 平均、運動量 (momentum) |
| $\nu$ | ニュー | 自由度、頻度 |
| $\pi$ | パイ | 円周率、確率分布 |
| $\rho$ | ロー | 相関係数、密度 |
| $\sigma$ | シグマ | 標準偏差、活性化関数 |
| $\tau$ | タウ | 時定数 |
| $\phi, \varphi$ | ファイ | 角度、写像 |
| $\omega$ | オメガ | 角周波数、重み |

大文字版もよく出る: $\Sigma$ (総和), $\Pi$ (総乗), $\Lambda$ (対角行列), $\Omega$ (大文字オメガ — 大きさ表記の O 記法と関連)。

## 2. Python の変数名にギリシャ文字を使う

Python 3 は Unicode 識別子に対応しているので、変数名にギリシャ文字をそのまま使える。

```python
# 直接ギリシャ文字を変数名に
α = 0.001       # 学習率
β₁ = 0.9        # Adam の β_1
β₂ = 0.999      # Adam の β_2
ε = 1e-8        # ゼロ割回避
θ = np.array([1.0, 2.0, 3.0])  # パラメータベクトル
σ = 0.5         # 標準偏差

print(f'学習率 α = {α}')
print(f'Adam β₁, β₂ = ({β₁}, {β₂})')
print(f'θ = {θ}')

# 注意: 読みやすさは賛否あり。一般的には英語綴り (alpha, beta, ...) を使う方が安全。
# パブリックコードでは英語、自分専用の探索 notebook ではギリシャ文字も OK
```

## 3. matplotlib のラベルで LaTeX を使う

グラフのタイトルや軸ラベルに数式を埋め込める (本リポジトリ全章で使用)。

ルール: `$...$` で囲み、中は **LaTeX 記法** で書く。

```python
# 正規分布のグラフ
μ_val, σ_val = 0, 1
x = np.linspace(-4, 4, 200)
y = (1 / (np.sqrt(2 * np.pi) * σ_val)) * np.exp(-(x - μ_val)**2 / (2 * σ_val**2))

fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(x, y, linewidth=2)
ax.set_title(r'正規分布 $\mathcal{N}(\mu, \sigma^2)$ with $\mu={}, \sigma={}$'.format(μ_val, σ_val))
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$p(x) = \frac{1}{\sqrt{2\pi}\sigma} \exp\!\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)$')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 4. LaTeX で頻出するギリシャ文字 (チートシート)

| LaTeX | 表示 |
|---|---|
| `\alpha \beta \gamma \delta` | $\alpha\,\beta\,\gamma\,\delta$ |
| `\epsilon \varepsilon \zeta \eta` | $\epsilon\,\varepsilon\,\zeta\,\eta$ |
| `\theta \vartheta \iota \kappa` | $\theta\,\vartheta\,\iota\,\kappa$ |
| `\lambda \mu \nu \xi` | $\lambda\,\mu\,\nu\,\xi$ |
| `\pi \rho \sigma \tau` | $\pi\,\rho\,\sigma\,\tau$ |
| `\phi \varphi \chi \psi \omega` | $\phi\,\varphi\,\chi\,\psi\,\omega$ |

大文字 (一部のみ — 小文字と同形のものは普通使われない):

| LaTeX | 表示 |
|---|---|
| `\Gamma \Delta \Theta \Lambda` | $\Gamma\,\Delta\,\Theta\,\Lambda$ |
| `\Sigma \Pi \Phi \Psi \Omega` | $\Sigma\,\Pi\,\Phi\,\Psi\,\Omega$ |

## 5. 機械学習論文で頻出する組合せ例

| 表現 | LaTeX | 意味 |
|---|---|---|
| $\theta^* = \arg\min_\theta L(\theta)$ | `\theta^* = \arg\min_\theta L(\theta)` | パラメータ最適解 |
| $\nabla_\theta L$ | `\nabla_\theta L` | パラメータ θ に関する勾配 |
| $\mathbb{E}_{x \sim p}[f(x)]$ | `\mathbb{E}_{x \sim p}[f(x)]` | 分布 p に従う x の期待値 |
| $\mathcal{N}(\mu, \sigma^2)$ | `\mathcal{N}(\mu, \sigma^2)` | 正規分布 |
| $\|\theta\|_2$ | `\|\theta\|_2` | L2 ノルム |

これらを「読めて、自分で書ける」 のが最終目標です。
